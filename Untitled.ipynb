```python
dic_scopus = {
    "DI": "DOI", 
    "AU": "Authors",
    "VL": "Volume",
    "CR": "References",
    "BP": "Page start",
    "PY": "Year",
    "SO": "Source title",
    "TI": "Title"
}
```

```python
for doi in df_scopus_minimal["DOI"]:
    row_wos = df_minimal[df_wos["DI"] == doi].iloc[0]
    row_scopus = df_scopus[df_scopus["DOI"] == doi].iloc[0]
    for c in row_wos.index:
        if c in dic_scopus:
            print(c)
            print("\nwos")
            print(row_wos[c])
            print("\nscopus")
            print(row_scopus[dic_scopus[c]])
            print()
    print()
```

```python
def au_to_wos(ex_au):
    names = ex_au.split(", ")
    new_names = []
    for name in names:
        n_name = name.replace(".", "")
        arr_n_name = n_name.split()
        n_name = " ".join(arr_n_name[:-1])  + ", " + arr_n_name[-1]
        new_names.append(n_name)
    return "; ".join(new_names)
au_to_wos("Biesma R.G., Pavlova M., van Merode G.G., Groot W.")
```

```python
# field tags: https://images.webofknowledge.com/images/help/WOS/hs_wos_fieldtags.html
```

```python
filepath = Path('D:reduced_scopus.txt')  

filepath.parent.mkdir(parents=True, exist_ok=True)  

rename_cols = {v:k for k, v in dic_scopus.items()}

df_scopus[dic_scopus["SO"]] = df_scopus[dic_scopus["SO"]].fillna("WO-SO")

df_scopus_filtered = df_scopus[~df_scopus[dic_scopus["SO"]].isna()]
df_scopus_filtered[dic_scopus["AU"]] = df_scopus_filtered[dic_scopus["AU"]].apply(au_to_wos)

df_scopus_filtered[dic_scopus.values()].rename(columns=rename_cols).to_csv(filepath, sep="\t") 
```
